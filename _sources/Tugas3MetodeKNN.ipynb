# Metode KNN


Metode KNN (K-Nearest Neighbor) adalah algoritma pembelajaran berbasis contoh. Metode KNN termasuk klasifikasi sangat baik. Dalam klasifikasi data terdapat dua proses yaitu learning dan classification. Teori ini digunakan untuk mengklasifikasikan berdasarkan jumlah dari K-neighbord terdekat.

```python
# import library yang diperlukan
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
type(iris)
```

```python
print(iris.feature_names)
```

```python
print(iris.target)
```

```python
print(iris.target_names)
```

```python
X = iris.data
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn import metrics
k_range = range(1,26)
scores = {}
score_list = []

for k in k_range:
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train, y_train)
  y_pred= knn.predict(X_test)
  scores[k]=metrics.accuracy_score(y_test, y_pred)
  score_list.append(metrics.accuracy_score(y_test, y_pred))
```

```python
%matplotlib inline

plt.plot(k_range, score_list)
plt.xlabel('value of K for KNN')
plt.ylabel('testing accuracy')
```

```python
classes = {0:'setosa', 1:'versicolor', 2:'virginica'}

new = [[3,2,4,4], [5,4,2,2]]
y_predict = knn.predict(new)

print(classes[y_predict[0]])
print(classes[y_predict[1]])
```
